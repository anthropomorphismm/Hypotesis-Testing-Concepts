```python
df_2n_Cycle = df[df['Education'] == '2n Cycle'] #Tes Hipotesis #Stratifikasi untuk tiap jenjang
df_Basic = df[df['Education'] == 'Basic']
df_Graduation = df[df['Education'] == 'Graduation']
df_Master = df[df['Education'] == 'Master']
df_PhD = df[df['Education'] == 'PhD']
```

```python
df_2n_Cycle = df_2n_Cycle[df_2n_Cycle['Income'].isna() == False] #Memfilter 'missing value' dari data
df_Basic = df_Basic[df_Basic['Income'].isna() == False]
df_Graduation = df_Graduation[df_Graduation['Income'].isna() == False]
df_Master = df_Master[df_Master['Income'].isna() == False]
df_PhD = df_PhD[df_PhD['Income'].isna() == False]
```

```python
H_0 = ('variasi jenjang dalam edukasi tidak memiliki pengaruh terhadap income') #menentukan hipotesis
H_1 = ('variasi jenjang dalam edukasi memiliki pengaruh signifikasn dalam income')
```

```python
anova = st.f_oneway(df_2n_Cycle['Income'], #P-value pada hasil tes
                    df_Basic['Income'],
                    df_Graduation['Income'],
                    df_Master['Income'],
                    df_PhD['Income'])
```

```python
p_value = anova.pvalue #hasil nilai tes p-value
p_value
```

```python
if p_value > 0.05: #pengambilan keputusan hipotesis
   print(H_0)
else:
   print(H_1)
```
